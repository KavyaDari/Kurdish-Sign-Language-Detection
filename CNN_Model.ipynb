```python
import os

DATA_DIR = r"C:\Datasets\kurdish_sign_dataset"
mapping_path = os.path.join(DATA_DIR, "class_mapping.txt")

class_folders = [f for f in os.listdir(DATA_DIR) if os.path.isdir(os.path.join(DATA_DIR, f))]
mapping = {}

for i, folder in enumerate(sorted(class_folders)):
    safe_name = f"class_{i+1}"
    old_path = os.path.join(DATA_DIR, folder)
    new_path = os.path.join(DATA_DIR, safe_name)
    os.rename(old_path, new_path)
    mapping[safe_name] = folder

with open(mapping_path, "w", encoding="utf-8") as f:
    for k, v in mapping.items():
        f.write(f"{k} -> {v}\n")

print("вң… All folders renamed safely.")
print("рҹ”Ө Mapping saved to:", mapping_path)
```

```python
import os

DATA_DIR = r"C:\Datasets\kurdish_sign_dataset"

def clean_names(directory):
    for root, dirs, files in os.walk(directory):
        for name in dirs + files:
            old_path = os.path.join(root, name)
            try:
                name.encode('utf-8')  # check if name is valid UTF-8
            except UnicodeEncodeError:
                new_name = name.encode('utf-8', 'ignore').decode('utf-8', 'ignore')
                new_path = os.path.join(root, new_name)
                os.rename(old_path, new_path)
                print(f"Renamed: {old_path} вҶ’ {new_path}")
            except Exception as e:
                print(f"Error with {old_path}: {e}")

clean_names(DATA_DIR)
print("вң… All filenames checked and cleaned.")
```

```python

# KURDISH SIGN LANGUAGE (KuSL2023) вҖ” CLEAN CNN MODEL

import os, time
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support


# CONFIGURATION

DATA_DIR = r"C:\Datasets\kurdish_sign_dataset"
IMG_SIZE = (64, 64)
BATCH_SIZE = 64
VALIDATION_SPLIT = 0.1
EPOCHS = 20

MODEL_PATH = r"C:\Datasets\models\kurdish_sign_cnn.keras"
REPORT_PATH = r"C:\Datasets\reports\kurdish_sign_report.txt"


# STEP 1 вҖ” Load Dataset

train_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=VALIDATION_SPLIT,
    subset="training",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    color_mode='rgb'
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=VALIDATION_SPLIT,
    subset="validation",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    color_mode='rgb'
)

class_names = train_ds.class_names
num_classes = len(class_names)
print(f"вң… Found {num_classes} classes: {class_names}")

# Normalize pixel values
def normalize(images, labels):
    images = tf.cast(images, tf.float32) / 255.0
    return images, labels

AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.map(normalize).shuffle(1000).prefetch(AUTOTUNE)
val_ds = val_ds.map(normalize).prefetch(AUTOTUNE)


# STEP 2 вҖ” Build CNN Model

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', padding='same', input_shape=(64,64,3)),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation='relu', padding='same'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(128, (3,3), activation='relu', padding='same'),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()


# STEP 3 вҖ” Train Model

start_time = time.time()
history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS)
train_time = time.time() - start_time

print(f"\nвҸұ Training completed in {train_time:.1f} seconds.")
os.makedirs(os.path.dirname(MODEL_PATH), exist_ok=True)
model.save(MODEL_PATH)
print(f"вң… Model saved at: {MODEL_PATH}")


# STEP 4 вҖ” Evaluate Model

y_true, y_pred = [], []
for batch_images, batch_labels in val_ds:
    preds = model.predict(batch_images)
    y_pred.extend(np.argmax(preds, axis=1))
    y_true.extend(batch_labels.numpy())

y_true, y_pred = np.array(y_true), np.array(y_pred)
acc = np.mean(y_true == y_pred)
precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted')

print(f"\nрҹ“Ҡ Accuracy: {acc:.4f}")
print(f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1: {f1:.4f}")
report = classification_report(y_true, y_pred, target_names=class_names, digits=4)
print("\nClassification Report:\n", report)

# Save report
os.makedirs(os.path.dirname(REPORT_PATH), exist_ok=True)
with open(REPORT_PATH, "w", encoding="utf-8") as f:
    f.write(f"Training time: {train_time:.2f}s\n")
    f.write(f"Accuracy: {acc:.4f}\nPrecision: {precision:.4f}\nRecall: {recall:.4f}\nF1: {f1:.4f}\n\n")
    f.write(report)
print(f"вң… Report saved at: {REPORT_PATH}")


# STEP 5 вҖ” Visualization

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.title("Training & Validation Accuracy")
plt.xlabel("Epochs"); plt.ylabel("Accuracy"); plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title("Training & Validation Loss")
plt.xlabel("Epochs"); plt.ylabel("Loss"); plt.legend()
plt.tight_layout()
plt.show()

cm = confusion_matrix(y_true, y_pred)
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
plt.figure(figsize=(12,10))
sns.heatmap(cm_norm, xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.title("Normalized Confusion Matrix")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.show()
```
